```python
"""Transposici√≥n con m√©todo:
Para cambiar filas por columnas
"""
import pandas as pd

# DataFrame de ejemplo
data = {
    "stu1": [90, 98, 100, 100],
    "stu2": [80, 89, 95, 90],
    "stu3": [70, 95, 100, 90]
}
df = pd.DataFrame(data, index=["math", "eng", "mus", "phy"])

# Transposici√≥n con el m√©todo
dff = df.transpose()   # o df.T
print(dff)
```

```python
"""
Transposici√≥n con atributo .T
Es un atajo de transpose(). Devuelve el mismo resultado que el ejercicio anterior.
"""
dff2 = df.T
print(dff2)
```

```python
"""Usar una columna como indice
Permite identificar filas por nombre en lugar que por n√∫mero
"""
# Crear un DataFrame con columna 'name'
data = {
    "name": ["stu1", "stu2", "stu3"],
    "math": [90, 80, 70],
    "eng": [98, 89, 95],
    "mus": [100, 95, 100],
    "phy": [100, 90, 90]
}
df = pd.DataFrame(data)

# Usar 'name' como √≠ndice
df2 = df.set_index("name")
print(df2)
```

```python
"""Multiindice
Crea un ejemplo jer√°rquico
"""
# Ejemplo multi√≠ndice
arrays = [
    ["stu1", "stu1", "stu2", "stu2"],
    ["math", "eng", "math", "eng"]
]
index = pd.MultiIndex.from_arrays(arrays, names=("student", "subject"))

scores = [90, 98, 80, 89]
df = pd.DataFrame(scores, index=index, columns=["score"])
print(df)
```

```python
"""Reindexaci√≥n
Se agregan nuevas filas y se rellenan con 0
"""
# Crear DataFrame simple
data = {"c1": [10, 20, 30]}
df = pd.DataFrame(data, index=["r0", "r1", "r2"])

# Reindexar con nuevos √≠ndices
df2 = df.reindex(["r0", "r1", "r2", "r3", "r4"])
print(df2)
```

```python
"""Resetear √≠ndice
Convierte el √≠ndice en una columna y vuelve al √≠ndice entero por defecto.

"""
# Resetear √≠ndice a num√©rico
df_reset = df2.reset_index()
print(df_reset)
```

```python
"""Ordenar √≠ndice
Permite organizar el DataFrame seg√∫n los √≠ndices.
"""
# Ordenar √≠ndice en orden descendente
df_sorted = df2.sort_index(ascending=False)
print(df_sorted)
```

```python
"""Ordenar por columna
Ordena el DataFrame en funci√≥n de los valores de una columna.
"""
# Crear un DataFrame con varias columnas
data = {"c1": [10, 40, 20], "c2": [30, 20, 50]}
df = pd.DataFrame(data, index=["r0", "r1", "r2"])

# Ordenar por valores de columna c1 (descendente)
df_sorted = df.sort_values(by="c1", ascending=False)
print(df_sorted)
```
