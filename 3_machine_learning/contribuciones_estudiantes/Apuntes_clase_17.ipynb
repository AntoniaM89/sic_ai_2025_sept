Uso del método jerárquico con el dataset Iris

```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()
iris
```

```python
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris
```

```python
from sklearn.cluster import AgglomerativeClustering
linkage=["complete", "average", "ward","single"]
for idx, i in enumerate(linkage):
    plt.figure(idx)
    hier=AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage=i)
    hier.fit(df_iris.iloc[:,2:4])
    plt.scatter(df_iris.iloc[:,2], df_iris.iloc[:,3], c=hier.labels_)
    plt.title("Clustering: "+ i)
    plt.xlabel('petal length')
    plt.ylabel('petal width')
    plt.show()
```

```python
from scipy.cluster import hierarchy
hierar=hierarchy.linkage(df_iris.iloc[:,2:4], 'complete')
plt.figure(figsize=(20,10))
dh=hierarchy.dendrogram(hierar)
```

```python
# El número se obtiene del gráfico anterior
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
labels = fcluster(hierar, 2.5, criterion='distance')
pd.Series(labels).value_counts()
```
