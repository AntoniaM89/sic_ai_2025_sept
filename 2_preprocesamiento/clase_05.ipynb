# C√≥mo cargar un CSV con membrete y comas en los n√∫meros en Pandas

Cuando un archivo CSV no est√° limpio, pueden aparecer errores como:

    ParserError: Error tokenizing data. C error: Expected X fields in line Y, saw Z

Esto ocurre generalmente por dos motivos:\
1. El archivo tiene filas de membrete antes de la cabecera real de
datos.\
2. Los n√∫meros vienen con comas (`112,002`) que Pandas interpreta como
separador de columnas.

------------------------------------------------------------------------

## Paso 1: Saltar las filas del membrete

En el ejemplo, los datos reales empiezan en la fila 7.\
Por eso usamos `skiprows=6` para ignorar las 6 primeras filas:

``` python
df = pd.read_csv("sd_082025.csv", skiprows=6)
```

------------------------------------------------------------------------

## Paso 2: Especificar el separador correcto

Muchos CSV en Chile usan punto y coma (`;`) como delimitador, porque la
coma se usa en los n√∫meros.\
Si no lo indicamos, Pandas se confunde.

``` python
df = pd.read_csv("sd_082025.csv", skiprows=6, sep=";", engine="python")
```

------------------------------------------------------------------------

## Paso 3: Interpretar las comas en los n√∫meros

Los montos vienen con formato `112,002`. Pandas puede entender esa coma
como separador de miles:

``` python
df = pd.read_csv(
    "sd_082025.csv",
    skiprows=6,
    sep=";",        # separador de columnas
    thousands=",",  # interpreta 112,002 como 112002
    engine="python"
)
```

------------------------------------------------------------------------

## Paso 4: Revisar las primeras l√≠neas del archivo

Si no est√°s seguro del separador, imprime unas filas:

``` python
with open("sd_082025.csv", "r", encoding="utf-8") as f:
    for i in range(15):
        print(f.readline())
```

Esto permite confirmar si el delimitador es `;`, `,` o tabulaci√≥n.

------------------------------------------------------------------------

## Resultado esperado

Despu√©s de estos pasos deber√≠as tener un DataFrame limpio con columnas
como:

-   Regi√≥n\
-   C√≥d Comuna\
-   Glosa Comuna\
-   N¬∞ Hombre\
-   Mto.Hombre\
-   N¬∞ Mujer\
-   Mto.Mujer\
-   N¬∞\
-   Monto m\$

````python
import pandas as pd


file_path = '/content/bitacora-vuelos (2).parquet'

# Intentar con pyarrow (el motor por defecto)
try:
    df = pd.read_parquet(file_path, engine='pyarrow')

    print("Le√≠do con pyarrow")
except Exception as e:
    print("Error con pyarrow:", e)
    # Intentar con fastparquet
    try:
        df = pd.read_parquet(file_path, engine='fastparquet')
        print("Le√≠do con fastparquet")
    except Exception as e2:
        print("Error con fastparquet:", e2)
        df = None

if df is not None:
    display(df.head(10))
else:
    print("No se pudo leer el archivo Parquet.")
````
