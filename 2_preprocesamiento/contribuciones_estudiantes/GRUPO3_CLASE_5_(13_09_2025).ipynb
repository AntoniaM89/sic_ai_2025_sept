# Operaciones en Serie

## Slides 266 - 267: Serie vs Número


> Series vs Numero

Una Serie es como una lista o columna de datos. Si le sumas, restas, multiplicas o divides un número a toda la serie, la operación se aplica automáticamente a cada elemento de esa serie (esto es lo que se llama vectorización).



```python
import numpy as np
import pandas as pd

student1 = pd.Series({'kor':100, 'eng':80, 'math':90})
#Se crea una Serie de Pandas a partir de un diccionario.
#La Serie tiene como índices: "kor", "eng", "math".
#Y como valores: 100, 80, 90.

student1
```

```python
percentage = student1 / 200
#Divide cada valor de la Serie entre 200.
#Pandas aplica la operación automáticamente a cada elemento.

percentage
```

- Una Serie de Pandas permite trabajar con datos de manera vectorizada.
- No hace falta usar bucles for para dividir o transformar cada valor: basta una sola operación.
- En el ejemplo, se transformaron las notas de un estudiante en porcentajes dividiendo todo entre 200.

## SLIDES 268 - 271: Serie vs Serie

> Serie vs. Serie

Aquí creamos dos Series de Pandas (que son como un diccionario + arreglo).

- student1 tiene las notas de un estudiante en korean (kor), english (eng), math.

- student2 tiene las notas de otro estudiante, pero el orden de las materias está cambiado.

```python
student1 = pd.Series({'kor':100, 'eng':80, 'math':90})
student2 = pd.Series({'math': 80, 'kor': 90, 'eng': 80})

addition = student1 + student2
subtraction = student1 - student2
multiplication = student1 * student2
division = student1 / student2

# Cuando haces operaciones como +, -, *, / entre dos pd.Series, Pandas alinea
# automáticamente los índices (las etiquetas, en este caso 'kor', 'eng', 'math') antes de operar.
```

```python
result= pd.DataFrame([addition,subtraction,multiplication,division],
                     index=['addition','subtraction','multiplication','division'])

# La lista [addition, subtraction, ...] se convierte en las filas del DataFrame.
# Con index=[...] le das nombres a esas filas.

result
```

```python
student1= pd.Series({'kor':np.nan,'eng':80,'math':90})
student2= pd.Series({'math':80,'kor':90})
addition =student1+student2
subtraction =student1-student2
multiplication =student1*student2
division =student1/student2
```

- student1 tiene kor = NaN.

- student1 tiene eng, pero student2 no tiene eng.

Cuando se hacen operaciones (+, -, *, /), Pandas alinea por los índices y luego aplica la operación.

- Si un índice existe en ambas Series → se hace la operación.

- Si falta en una de ellas → se devuelve NaN (porque el resultado es indefinido).

- Si uno de los valores es NaN → también se devuelve NaN.

```python
result = pd.DataFrame([addition,subtraction,multiplication,division],
                      index=['addition','subtraction','multiplication','division'])
result
```

## SLIDES 272 - 273: Método de Operación

> Método de Operación


```python
student1 = pd.Series({'kor': np.nan, 'eng': 80, 'math':90})
student2 = pd.Series({'math': 80, 'kor': 90})

sr_add = student1.add(student2, fill_value=0) # Esto trata los valores inexistentes como 0 en la suma.
sr_sub = student1.sub(student2, fill_value=0) # Esto trata los valores inexistentes como 0 en la substracción.
sr_mul = student1.mul(student2, fill_value=1) # Esto trata los valores inexistentes como 1 en la multiplicación.
sr_div = student1.div(student2, fill_value=1) # Esto trata los valores inexistentes como 1 en la división.

# Si queremos evitar esos NaN, podemos usar métodos como .add(), .sub(), .mul() y .div()
# Con el parámetro fill_value, reemplazamos los faltantes antes de operar.
```

```python
result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div],
                      index=['add', 'sum', 'mul', 'div'])

# El valor que uses en fill_value importa mucho:
#     En suma/resta → usar 0 es lo natural.
#     En multiplicación/división → usar 1 es más seguro (si usas 0, anulas todo).
# Esto hace que tus resultados sean más “limpios” y sin NaN.
```
