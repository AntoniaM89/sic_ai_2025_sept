###Fusión/“merge” (unir por una(s) columna(s) clave)

Tipos de unión que debes saber:
*   inner: solo las filas que coinciden en ambas tablas.
*   left: todas las filas de la izquierda, y lo que coincida de la derecha.
*   right: todas las filas de la derecha, y lo que coincida de la izquierda.
*   outer (externa completa): todas las filas de ambas, coincidan o no.

```python
import pandas as pd
```

```python
# Tabla de estudiantes
df_est = pd.DataFrame({
    "id": [1, 2, 3],
    "nombre": ["Ana", "Ben", "Cara"]
})

# Tabla de notas
df_notas = pd.DataFrame({
    "id": [1, 3, 4],
    "nota": [6.0, 5.5, 4.0]
})
```

```python
df_est
```

```python
df_notas
```

```python
#Inner
```

```python
pd.merge (df_est,df_notas,on="id",how="inner")
```

```python
pd.merge (df_est,df_notas,left_on='id',right_on='id',how='left')
```

###Enlace/“concat”

*   Vertical (una debajo de la otra): coincide por nombres de columna.
*  Horizontal (lado a lado): coincide por índices.

```python
import pandas as pd

enero  = pd.DataFrame({"dia":[1,2], "ventas":[10,12]})
febrero= pd.DataFrame({"dia":[1,2], "ventas":[9,11]})
```

```python
pd.concat([enero,febrero],sort= True) #vertical
```

```python
pd.concat([enero,febrero],axis=1,sort= True) #horinzontal

```
