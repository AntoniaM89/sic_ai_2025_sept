### importar librerias

```python
import pandas as pd
import seaborn as sns
```

### Desde la funcion load_dataset importamos coleccion de datos titanic

```python
titanic = sns.load_dataset("titanic")
titanic
```

### con "LOC" -> seleccionamos las columnas edad y tarifa, ":" hace referencia a todos los datos

```python

df = titanic.loc[:, ['age', 'fare']]
df.head()
```

### ejemplo de filtro en .LOC

```python

df = titanic.loc[titanic['age'] > 30, ['age', 'fare']]
df.head()
```

### Adicion

```python
addition = df+10
addition.head()
```

### subtraccion

```python
subtract = addition-df
subtract.head()
```
