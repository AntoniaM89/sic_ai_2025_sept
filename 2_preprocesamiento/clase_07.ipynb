# Colores en histogramas de Matplotlib

La función `matplotlib.pyplot.hist` permite personalizar los colores de
las barras del histograma (también llamadas *bins*). A continuación se
explican los parámetros y técnicas más comunes.

------------------------------------------------------------------------

## 1. `color`

Controla el color de **relleno** de las barras.

-   Puede ser:
    -   Un nombre de color estándar: `"red"`, `"blue"`, `"green"`,
        `"skyblue"`, etc.
    -   Un código **HEX**: `"#FF5733"`.
    -   Un valor RGB o RGBA en tupla: `(0.2, 0.5, 0.8)` o
        `(0.2, 0.5, 0.8, 0.6)` (el último valor es la transparencia).
    -   Una **lista de colores**, donde cada bin recibe un color
        distinto.

``` python
plt.hist(data, bins=10, color="skyblue")
plt.hist(data, bins=10, color=["red", "blue", "green"])
```

------------------------------------------------------------------------

## 2. `edgecolor`

Define el color de los **bordes** de cada barra.

``` python
plt.hist(data, bins=10, color="lightblue", edgecolor="black")
```

------------------------------------------------------------------------

## 3. `linewidth` o `lw`

Controla el grosor de la línea del borde. Útil en combinación con
`edgecolor`.

``` python
plt.hist(data, bins=10, edgecolor="black", linewidth=2)
```

------------------------------------------------------------------------

## 4. `alpha`

Ajusta la **transparencia** de las barras.\
El valor va entre `0` (completamente transparente) y `1` (completamente
opaco).

``` python
plt.hist(data, bins=10, color="orange", alpha=0.5)
```

------------------------------------------------------------------------

## 5. Uso de colormaps (avanzado)

Es posible asignar un color distinto a cada barra de acuerdo con la
altura del bin, usando **colormaps** de Matplotlib.

``` python
import matplotlib.cm as cm

counts, bins, patches = plt.hist(data, bins=10)

for count, patch in zip(counts, patches):
    plt.setp(patch, "facecolor", cm.viridis(count / max(counts)))
```

En este ejemplo, el mapa de colores `viridis` se aplica de manera que
las barras más altas se colorean diferente a las más bajas.

------------------------------------------------------------------------

## Resumen

-   `color`: define el relleno de las barras.
-   `edgecolor`: define el color de los bordes.
-   `linewidth`: ajusta el grosor del borde.
-   `alpha`: controla la transparencia.
-   Colormaps: permiten aplicar colores dinámicos basados en la
    frecuencia de cada bin.

Con estos parámetros es posible lograr desde un histograma básico hasta
visualizaciones más elaboradas y personalizadas.

# Colormaps en Seaborn

Seaborn permite usar **colormaps** (mapas de color) tanto propios como
los heredados de Matplotlib. Estos se utilizan en funciones que aceptan
el parámetro `cmap`, por ejemplo en `sns.heatmap`.

------------------------------------------------------------------------

## 1. Paletas de Seaborn convertidas en `cmap`

En muchas funciones de Seaborn puedes usar las paletas como colormaps si
agregas `as_cmap=True`.

Ejemplos:

``` python
sns.color_palette("rocket", as_cmap=True)
sns.color_palette("mako", as_cmap=True)
sns.color_palette("flare", as_cmap=True)
sns.color_palette("crest", as_cmap=True)
sns.color_palette("viridis", as_cmap=True)  # wrapper sobre Matplotlib
sns.cubehelix_palette(as_cmap=True)
```

------------------------------------------------------------------------

## 2. Generadores de paletas en Seaborn que aceptan `as_cmap=True`

-   `sns.cubehelix_palette(as_cmap=True)` → colormap tipo *cubehelix*.
-   `sns.light_palette("seagreen", as_cmap=True)` → colormap claro a
    partir de un color base.
-   `sns.dark_palette("purple", as_cmap=True)` → colormap oscuro a
    partir de un color base.
-   `sns.blend_palette(["red", "blue"], as_cmap=True)` → colormap
    mezclando dos o más colores.

Ejemplo:

``` python
sns.heatmap(data, cmap=sns.light_palette("navy", as_cmap=True))
```

------------------------------------------------------------------------

## 3. Paletas predefinidas de Matplotlib (también válidas en Seaborn)

Seaborn hereda todos los colormaps de Matplotlib, por lo que puedes usar
directamente:

-   **Perceptualmente uniformes**: `"viridis"`, `"plasma"`, `"inferno"`,
    `"magma"`, `"cividis"`.
-   **Secuenciales**: `"Greens"`, `"Blues"`, `"Oranges"`, `"Purples"`.
-   **Divergentes**: `"coolwarm"`, `"bwr"`, `"seismic"`.
-   **Cualitativas**: `"tab10"`, `"Set1"`, `"Set2"`, `"Pastel1"`,
    `"Accent"`.

``` python
sns.heatmap(data, cmap="viridis")
sns.heatmap(data, cmap="coolwarm")
```

------------------------------------------------------------------------

## Resumen

-   Con `sns.color_palette(..., as_cmap=True)` puedes convertir
    cualquier paleta de Seaborn en un colormap.
-   Con `sns.cubehelix_palette`, `sns.light_palette`,
    `sns.dark_palette`, `sns.blend_palette` puedes generar colormaps
    personalizados.
-   Todos los colormaps de Matplotlib también están disponibles en
    Seaborn.

````python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

N = 50

fig = make_subplots(rows=2, cols=2,
                    specs=[[{'is_3d': True}, {'is_3d': True}],
                           [{'is_3d': True}, {'is_3d': True}]],
                    print_grid=False)
for i in [1,2]:
    for j in [1,2]:
        fig.add_trace(
            go.Mesh3d(
                x=(60*np.random.randn(N)),
                y=(25*np.random.randn(N)),
                z=(40*np.random.randn(N)),
                opacity=0.5,
              ),
            row=i, col=j)

fig.update_layout(width=700, margin=dict(r=10, l=10, b=10, t=10))
# fix the ratio in the top left subplot to be a cube
fig.update_layout(scene_aspectmode='cube')
# manually force the z-axis to appear twice as big as the other two
fig.update_layout(scene2_aspectmode='manual',
                  scene2_aspectratio=dict(x=1, y=1, z=2))
# draw axes in proportion to the proportion of their ranges
fig.update_layout(scene3_aspectmode='data')
# automatically produce something that is well proportioned using 'data' as the default
fig.update_layout(scene4_aspectmode='auto')
fig.show()
````
